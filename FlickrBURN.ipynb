```python
import pandas as pd
df = pd.read_json('HoR44.json')
```

```python
df.head()
```

```python
farmid = 'sdfd'
def urler(df):
    root = "https://farm{farmid}.staticflickr.com/{serverid}/{id}_{secret}.jpg".format(\
     farmid=df['farm'],serverid=df['server'],id=df['id'],secret=df['secret'])
    return root
```

```python
df['url'] = df.apply(urler,axis=1)
```

```python
def splitter(row):
    return row.split("_")[0],row.split("_")[1],row.split("_")[3]

three = df['title'].apply(splitter)
```

```python
foo = lambda x: pd.Series([i for i in x.split('_')])
rev = df['title'].apply(foo).rename(columns={0:'mp',1:'parliament',2:'skip',3:'page'})
df = pd.concat([df,rev],axis=1).sort_values(['title','page'])
df
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('http://www.aph.gov.au/Senators_and_Members/Members/Register/Previous_Parliaments/44P_Members_Interest_Statements')
```

```python
soup = BeautifulSoup(page.content)
trs = soup.find_all("tr")
rows = []
for tr in trs:
    tds = list(tr.find_all('td'))
    if len(tds)==3:
        rows.append([tds[0].text.strip(),tds[1].text.strip(),tds[2].a.get('href').split(".")[0].split("/")[-1].split("_")[0]])
```

```python
dfr = pd.DataFrame(rows,columns=['updated','fullmp','mp'])
dfr.head()
```
